```python
# @title Run this code cell and select a demo to download!

!git clone --depth 1 https://github.com/JohnSnowLabs/spark-nlp.git

import os, shutil

path = "/content/spark-nlp/examples/demos/streamlit/"
folders = sorted(next(os.walk(path))[1])
print(f"Available Demos:\n" + "\n".join(f"{i + 1}. {folder}" for i, folder in enumerate(folders)))

choice = sorted(folders)[int(input("Enter the number of the Demo you want to download: ")) - 1]
dest = f"./{choice}"

if not os.path.exists(dest):
    shutil.copytree(os.path.join(path, choice), dest)
    print(f"Demo '{choice}' copied successfully.")
else:
    print(f"The Demo '{choice}' already exists in this directory. Skipping download.")

shutil.rmtree('/content/spark-nlp')
```
